```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
s = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2015', periods=1000))
```

```python
s = s.cumsum() # 숫자를 모두 더해줌
s.plot()
```

```python
r = s.rolling(window=30)
r.mean() #평균값 출력
# rolling(); window의 사이즈만큼 값을 이동 시켜줌
```

```python
# 이동평균 moving average
s.plot(style='k--')
r.mean().plot(style='k') # rolling -> 평균값 -> plot
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=pd.date_range('1/1/2015', periods=1000),
                  columns=['A', 'B', 'C', 'D'])
df = df.cumsum()
df.plot()
```

```python
df.rolling(window=60).sum().plot(subplots=True)
```

```python
df.rolling(window=len(df), min_periods=1).mean().plot()
```

```python
# 값을 누적해서 확장
df.expanding(min_periods=1).mean().plot()
```
