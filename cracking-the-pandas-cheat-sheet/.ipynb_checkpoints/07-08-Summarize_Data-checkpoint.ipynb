### apply(function)
---
* 데이터 전처리에서 사용되는 함수
* 특정 값까지 잘라오는 형태로 많이 사용
* 파라미터로 직접 lambda함수를 사용

```python
def smp(x):
    # 뒤에서 3번째까지의 문자를 가져오는 함수
    x = x[-3:]
    return x
```

```python
df['species_3'] = df['species'].apply(lambda x : x[:3])
```

```python
df['species-3'] = df['species'].apply(smp)
```
